### Task 1

Известно, что генеральная совокупность распределена нормально
со средним квадратическим отклонением, равным 16.
Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95,
если выборочная средняя M = 80, а объем выборки n = 256.

```python
import numpy as np
```

Так как известно среднее квадратическое отклонение генеральной совокупности, то для расчёта доверительного интервала будем использовать Z-критерий. По таблице Z-значений определим Z<sub>α/2</sub> для 95% доверительного интервала. Z<sub>α/2</sub> = 1,96. По формуле определим доверительный интервал:

```python
x_mean = 80
z_a2 = 1.96
n = 256
sigma = 16

def conf_int(x, n, z, sigma):
    left_part = x - z * sigma / (n ** 0.5)
    right_part = x + z * sigma / (n ** 0.5)
    return f'{left_part:.2f}; {right_part:.2f}'

ci = conf_int(x_mean, n, z_a2, sigma)
print(f'Доверительный интервал для оценки математического ожидания: [{ci}]')
```

Ответ: Доверительный интервал для оценки математического ожидания: [78.04; 81.96].

### Task 2

В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью,
получены опытные данные:
6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1
Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей,
оценить истинное значение величины X при помощи доверительного интервала, покрывающего это
значение с доверительной вероятностью 0,95.

Так как неизвестно среднее квадратичное отклонение генеральной совокупности, то для расчёта доверительного интервала будем использовать t-критерий.

Определим несмещённое среднее квадратическое отклонение выборки.

```python
ar = np.array([6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1])
ar_std = np.std(ar, ddof=1)
print(f'Несмещённое среднее квадратическое отклонение выборки — {ar_std:.3f}')
```

Определим t по таблице для построения 95% доверительного интервала, если объем выборки n = 10. Число степеней свободы v = n - 1 = 9.

```python
x_mean = np.mean(ar)
n = 10
t_a2 = 2.262
sigma_n = 0.451

ci = conf_int(x_mean, n, t_a2, sigma_n)
print(f'Доверительный интервал для истинного значения X: [{ci}]')
```

Ответ: доверительный интервал для истинного значения X: [6.27; 6.91].

### Task 3

Проведите тест гипотезы. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр
оказался равным 17.5 мм, а дисперсия известна и равна 4 кв. мм.

H<sub>0</sub>: μ = μ<sub>0</sub><br>
H<sub>1</sub>: μ > μ<sub>0</sub>

Так как известна дисперсия, то для проверки гипотезы будем использовать Z-критерий. По таблице Z-значений определим Z<sub>α</sub> с уровнем статистической значимости α = 0.05. Z<sub>α</sub> = 1,645.

Вычислим Z-значение:

```python
x = 17.5
mu = 17
n = 100
sigma = 2

z = (x - mu) / (sigma / (n ** 0.5))
print(f'Расчётное Z-значение — {z:.3f}')
```

Т.к. расчётное Z-значение больше табличного, значит оно попадает в область принятия гипотезы H<sub>1</sub>.

Ответ: гипотеза H<sub>0</sub> не верна при α = 0.05.

### Task 4

Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья составляет 200 г.
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
Известно, что их веса распределены нормально.
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%? (Провести двусторонний тест.)

H<sub>0</sub>: μ = μ<sub>0</sub><br>
H<sub>1</sub>: μ ≠ μ<sub>0</sub>

Так как неизвестно среднее квадратичное отклонение генеральной совокупности, то для расчёта доверительного интервала будем использовать t-критерий. Уровень значимости α = 1 - 0.99 = 0.01. Определим несмещённое среднее квадратическое отклонение выборки.

```python
ar = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
ar_std = np.std(ar, ddof=1)
print(f'Несмещённое среднее квадратическое отклонение выборки — {ar_std:.3f}')
```

Определим t по таблице для α/2 = 0.005, если объем выборки n = 10. Число степеней свободы v = n - 1 = 9. t<sub>α/2</sub> = 3,25.

Вычислим наблюдаемое t-значение:

```python
x_mean = np.mean(ar)
mu = 200
n = 10
t_a2 = 3.25
sigma_n = 4.453

tn = (x_mean - mu) / (sigma_n / (n ** 0.5))
print(f'Наблюдаемое t-значение — {tn:.3f}')
```

Т.к. наблюдаемое t-значение меньше табличного, значит оно попадает в область принятия гипотезы H<sub>0</sub>.

Ответ: гипотеза H<sub>0</sub> верна при α = 0.01.
