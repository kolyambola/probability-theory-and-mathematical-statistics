### Task 1

Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
import numpy as np
from math import factorial
```

Создадим массив из выборки:

```python
x = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
x.shape
```

Определим среднее арифметическое значение:

```python
x_average = sum(x) / len(x)
x_average
```

Вычислим среднее квадратичное отклонение:

```python
x_std = np.sqrt(((x - x_average) ** 2).sum() / len(x))
x_std
```

Определим смещённую дисперсию по выборке:

```python
x_baised_var = ((x - x_average) ** 2).sum() / len(x)
x_baised_var
```

Определим несмещённую дисперсию по выборке:

```python
x_unbaised_var = ((x - x_average) ** 2).sum() / (len(x) - 1)
x_unbaised_var
```

Ответ: среднее арифметическое — 65.3; среднее квадратичное отклонение — 30.82; смещённая оценка дисперсии — 950.11; несмещённая оценка дисперсии — 1000.12.

### Task 2

В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

Для начала определим общее число исходов при вытаскивании мячей из каждого ящика.

```python
def combinations(n, k):
    c = factorial(n) / (factorial(k) * factorial(n - k))
    return c
```

Для первого ящика:

```python
n1 = int(combinations(8, 2))
print(f'{n1} сочетаний из 2 мячей')
```

Для второго ящика:

```python
n2 = int(combinations(12, 4))
print(f'{n1} сочетаний из 4 мячей')
```

Для данного варианта существует 3 события: а) из первого ящика извлекли два белых мяча, из второго — один белый и три не белых; б) из первого ящика извлекли один белый мяч и один не белый, из второго — два белых и два не белых; в) из первого ящика не извлекли ни одного белого мяча, из второго — три белых и один не белый.

Определим благоприятствующее число исходов для каждого события:

```python
m_aa = int(combinations(5, 2))
m_ba = int(combinations(5, 1) * combinations(7, 3))
print(f'а) Число благоприятствующих исходов события A — {m_aa}')
print(f'   Число благоприятствующих исходов события В — {m_ba}\n')

m_ab = int(combinations(5, 1) * combinations(3, 1))
m_bb = int(combinations(5, 2) * combinations(7, 2))
print(f'б) Число благоприятствующих исходов события A — {m_ab}')
print(f'   Число благоприятствующих исходов события В — {m_bb}\n')

m_ac = int(combinations(3, 2))
m_bc = int(combinations(5, 3) * combinations(7, 1))
print(f'в) Число благоприятствующих исходов события A — {m_ac}')
print(f'   Число благоприятствующих исходов события В — {m_bc}')
```

Вероятность, что сложится одна из этих комбинаций будет равна сумме вероятностей этих событий.

```python
p = (m_aa/n1)*(m_ba/n2) + (m_ab/n1)*(m_bb/n2) + (m_ac/n1)*(m_bc/n2)
print(f'Вероятность, что извлечённые из ящиков три мяча будут белыми — {p:.4f}')
```

Ответ: вероятность того, что 3 мяча белые — 0.3686.

### Task 3

На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

Пусть событие А — попадание в мишень, а события В1, В2, В3 — выстрел произведён 1-м, 2-м, 3-м спортсменом соответственно. Вероятности выстрела каждым спорстменом равны, так как выстрел один, а спортсменов три. Вероятность будет равна P(B1) = P(B2) = P(B3) = 1/3.

Определим вероятность наступления события A:

```python
p_b = 1/3
p_a = p_b*0.9 + p_b*0.8 + p_b*0.6
print(f'Полная вероятность наступления события A — {p_a:.4f}')
```

По формуле Байеса определим вероятность того, что выстрел произведён одним из спортсменов:

```python
p_1 = p_b * 0.9 / p_a
print(f'Вероятность, что выстрел произведён первым спортсменом — {p_1:.4f}')

p_2 = p_b * 0.8 / p_a
print(f'Вероятность, что выстрел произведён вторым спортсменом — {p_2:.4f}')

p_3 = p_b * 0.6 / p_a
print(f'Вероятность, что выстрел произведён третьим спортсменом — {p_3:.4f}')
```

Ответ: а) вероятность, что выстрел произведён первым спортсменом — 0.3913; б) вероятность, что выстрел произведён вторым спортсменом — 0.3478; в) вероятность, что выстрел произведён третьим спортсменом — 0.2609.

### Task 4

В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

Пусть n — число студентов факультетов А и В. Тогда число студентов факультета С — 2n. Введём события S(A), S(B), S(C), что сессию сдаёт студент факультетов A, B, C соответсвенно. Полная группа событий: P(S(A)) + P(S(B)) + P(S(C)) = 1. 1/n + 1/n + 1/2n = 1, n = 2.5.

Пусть событие D — сессия сдана. Вероятность наступления данного события:

```python
n = 2.5
p_d = (1/n)*0.8 + (1/n)*0.7 + (1/(2*n))*0.9
print(f'P(D) = {p_d:.4f}')
```

По формуле Байеса вычислим вероятность того, на каком факультете учится студент, сдавший сессию.

```python
p_da = (1/n)*0.8/p_d
print(f'Вероятность, что студент учится на факультете A — {p_da:.4f}')

p_db = (1/n)*0.7/p_d
print(f'Вероятность, что студент учится на факультете B — {p_db:.4f}')

p_dc = (1/(2*n))*0.9/p_d
print(f'Вероятность, что студент учится на факультете C — {p_dc:.4f}')
```

Ответ: а) Вероятность, что студент учится на факультете A — 0.4103; б) Вероятность, что студент учится на факультете B — 0.3590; в) Вероятность, что студент учится на факультете C — 0.2308.

### Task 5

Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

а) Выход из строя всех деталей — это совместные события, значит вероятности этих событий перемножаются.

```python
p = 0.1 * 0.2 * 0.25
print(f'Вероятность, что в первый месяц выйдут из строя все детали — {p:.4f}')
```

б) Чтобы из строя вышли только две детали, нужно выполнение одной из следующих комбинаций событий: 1) вышли из строя 1 и 2 детали, 3 — не вышла; 2) вышли из строя 1 и 3 детали, 2 — не вышла; 3) 2 и 3 детали вышли из строя, 1 — не вышла. Вероятности данных комбинаций будут складываться, так как события несовместные.

```python
p = 0.1*0.2*(1-0.25) + 0.1*0.25*(1-0.2) + 0.2*0.25*(1-0.1)
print(f'Вероятность, что в первый месяц выйдут из строя только две детали — {p:.4f}')
```

в) Рассмотрим событие — не вышло из строя ни одной детали. Определим вероятность данного события. А для нахождения исходного события вычтем событие "не вышло из строя ни одной детали" из единицы.

```python
p = 1 - 0.9*0.8*0.75
print(f'Вероятность, что в первый месяц выйдет из строя хотя бы одна деталь — {p:.4f}')
```

г) Данное событие произойдёт при наступлении события "из строя выйдет одна деталь" или "из строя выйдут две детали". Вероятность требуемого события будет равна сумме этих событий. 

Чтобы из строя вышла одна деталь, нужно выполнение одной из следующих комбинаций событий: 1) вышла из строя 1 деталь, 2, 3 — не вышли; 2) вышла из строя 2 деталь, 1 и 3 — не вышли; 3) 3 деталь вышла из строя, 1 и 2 — не вышли. Вероятности данных комбинаций будут складываться, так как события несовместные.

```python
p_1 = 0.1*(1-0.2)*(1-0.25) + 0.2*(1-0.1)*(1-0.25) + 0.25*(1-0.1)*(1-0.2)
print(f'Верояность, что из строя выйдет 1 деталь — {p_1:.4f}')
```

Определим вероятность того, что из строя выйдут от одной до двух деталей:

```python
p = p_1 + 0.08
print(f'Вероятность, что в первый месяц выйдут из строя от одной до двух деталей — {p:.4f}')
```

Ответ: а) Вероятность, что в первый месяц выйдут из строя все детали — 0.0050; б) Вероятность, что в первый месяц выйдут из строя только две детали — 0.0800; в) Вероятность, что в первый месяц выйдет из строя хотя бы одна деталь — 0.4600; г)Вероятность, что в первый месяц выйдут из строя от одной до двух деталей — 0.4550.
